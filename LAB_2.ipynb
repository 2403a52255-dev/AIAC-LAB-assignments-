Task-1: write a python code that reads a list of numbers and calculates the mean,median and mode of that numbers.

```python
def analyze_numbers(numbers):
  """
  Calculates the mean, minimum, and maximum of a list of numbers.

  Args:
    numbers: A list of numbers.

  Returns:
    A tuple containing the mean, minimum, and maximum values.
    Returns None if the list is empty.
  """
  if not numbers:
    return None

  mean = sum(numbers) / len(numbers)
  minimum = min(numbers)
  maximum = max(numbers)

  return mean, minimum, maximum

# Example usage:
my_list = [10, 20, 30, 40, 50]
result = analyze_numbers(my_list)

if result:
  mean_value, min_value, max_value = result
  print(f"List: {my_list}")
  print(f"Mean: {mean_value}")
  print(f"Minimum: {min_value}")
  print(f"Maximum: {max_value}")
else:
  print("The list is empty.")

empty_list = []
result_empty = analyze_numbers(empty_list)

if result_empty:
  mean_value, min_value, max_value = result_empty
  print(f"\nList: {empty_list}")
  print(f"Mean: {mean_value}")
  print(f"Minimum: {min_value}")
  print(f"Maximum: {max_value}")
else:
  print("\nThe list is empty.")
```

### Armstrong Number Checker

This section contains a Python function to determine if a given number is an Armstrong number.

**What is an Armstrong Number?**

An Armstrong number (also known as a narcissistic number, pluperfect digital invariant (PPDI), or plus perfect number) is a number that is the sum of its own digits each raised to the power of the number of digits.

For example, 153 is an Armstrong number because:
$1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153$

```python
def is_armstrong_number(number):
  """
  Checks if a number is an Armstrong number.

  Args:
    number: An integer.

  Returns:
    True if the number is an Armstrong number, False otherwise.
  """
  # Convert the number to a string to easily access its digits
  num_str = str(number)
  num_digits = len(num_str)
  sum_of_powers = 0

  # Calculate the sum of each digit raised to the power of the number of digits
  for digit in num_str:
    sum_of_powers += int(digit) ** num_digits

  # Check if the sum is equal to the original number
  return sum_of_powers == number

# Example usage:
num1 = 153
num2 = 123

print(f"{num1} is an Armstrong number: {is_armstrong_number(num1)}")
print(f"{num2} is an Armstrong number: {is_armstrong_number(num2)}")
```

 EXPLANATION :This code checks if a number is an "Armstrong number". An Armstrong number is a number where the sum of its digits, each raised to the power of the total number of digits, equals the original number. The code converts the number to text to count the digits and then checks if the sum of each digit raised to that power matches the original number.

TASK-3:write a python function to check whether given number is palindrome or not

```python
def is_palindrome(number):
  """
  Checks if a number is a palindrome.

  A palindrome number reads the same forwards and backward.

  Args:
    number: An integer.

  Returns:
    True if the number is a palindrome, False otherwise.
  """
  # Convert the number to a string
  num_str = str(number)

  # Reverse the string
  reversed_str = num_str[::-1]

  # Check if the original string is equal to the reversed string
  return num_str == reversed_str

# Example usage:
num1 = 121
num2 = 123

print(f"{num1} is a palindrome: {is_palindrome(num1)}")
print(f"{num2} is a palindrome: {is_palindrome(num2)}")
```

 TASK-4:write a python funtion to get sum of first n natural numbers. n should be taken from console

```python
def sum_of_first_n_natural_numbers(n):
  """
  Calculates the sum of the first n natural numbers.

  Args:
    n: An integer representing the number of natural numbers to sum.

  Returns:
    The sum of the first n natural numbers.
    Returns None if n is not a positive integer.
  """
  if not isinstance(n, int) or n <= 0:
    print("Please provide a positive integer for n.")
    return None
  else:
    # The formula for the sum of the first n natural numbers is n * (n + 1) / 2
    return n * (n + 1) // 2 # Use integer division

# Get input from the user
try:
  n_input = int(input("Enter a positive integer (n): "))
  result = sum_of_first_n_natural_numbers(n_input)

  if result is not None:
    print(f"The sum of the first {n_input} natural numbers is: {result}")

except ValueError:
  print("Invalid input. Please enter an integer.")
```

EXPLANATION:the program starts a function called sum_of_natural_numbers that takes one number, n. *If the number n you give is less than 1, it tells you to enter a positive number. *If n is 1 or more, it uses a quick math trick (n * (n + 1) // 2) to find the sum. *The part outside the function asks you to enter a positive number, uses the function to get the sum, and then prints the answer.

 TASK-5:write a Python program to calculate the sum of
odd numbers and even numbers in a given tupletas,tuple should be taken from console.

```python
def sum_odd_even_in_tuple(numbers_tuple):
  """
  Calculates the sum of odd and even numbers in a tuple.

  Args:
    numbers_tuple: A tuple of numbers.

  Returns:
    A tuple containing the sum of odd numbers and the sum of even numbers.
  """
  sum_odd = 0
  sum_even = 0

  for number in numbers_tuple:
    if isinstance(number, (int, float)): # Check if the element is a number
      if number % 2 == 0:
        sum_even += number
      else:
        sum_odd += number
    else:
      print(f"Skipping non-numeric element: {number}")

  return sum_odd, sum_even

# Get input from the user
input_str = input("Enter a tuple of numbers (e.g., 1, 2, 3, 4, 5): ")

# Convert the input string to a tuple of numbers
try:
  # Using eval is generally not recommended for untrusted input,
  # but for this specific educational context with the user providing input directly,
  # it's a straightforward way to parse a tuple from a string.
  # For production code, consider safer parsing methods.
  numbers_tuple = eval(input_str)

  if isinstance(numbers_tuple, tuple):
    odd_sum, even_sum = sum_odd_even_in_tuple(numbers_tuple)
    print(f"Original tuple: {numbers_tuple}")
    print(f"Sum of odd numbers: {odd_sum}")
    print(f"Sum of even numbers: {even_sum}")
  else:
    print("Invalid input. Please enter a valid tuple format.")

except (SyntaxError, NameError):
  print("Invalid input. Please enter a valid tuple format.")
```

EXPLANATION:This code defines a function sum_odd_even_in_tuple that takes a tuple of numbers as input. It initializes two variables, sum_odd and sum_even, to zero. It then loops through each element in the input tuple.

Inside the loop, it first checks if the element is an integer or a float to ensure it's a number. If it is a number, it checks if the number is even (number % 2 == 0) or odd. If it's even, it adds the number to sum_even; otherwise, it adds it to sum_odd. If the element is not a number, it prints a message indicating that it's skipping that element.

After the loop finishes, the function returns the calculated sum_odd and sum_even as a tuple.

The code then prompts the user to enter a tuple of numbers as a string (e.g., "1, 2, 3"). It uses eval() to convert this input string into a Python tuple. While eval() works here for this specific educational context, it's important to be aware that using eval() with untrusted user input can be dangerous in real-world applications.

Finally, it checks if the input was successfully converted to a tuple. If it is, it calls the sum_odd_even_in_tuple function with the user's tuple and prints the original tuple, the sum of odd numbers, and the sum of even numbers. If the input was not a valid tuple format, it prints an error message.
